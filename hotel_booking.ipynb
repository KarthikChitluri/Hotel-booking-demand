#### 1. Price variations per night over the year

```python
df_resort = df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == 0)]   # only for not canceled bookings
df_city = df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == 0)]
```

```python
resort_hotel_price = df_resort.groupby(['arrival_date_month'])['adr'].mean().reset_index()
resort_hotel_price
```

```python
city_hotel_price = df_city.groupby(['arrival_date_month'])['adr'].mean().reset_index()
city_hotel_price
```

```python
total_hotels_price = resort_hotel_price.merge(city_hotel_price, on = 'arrival_date_month')
total_hotels_price.columns = ['month', 'resort_hotel_price', 'city_hotel_price']
total_hotels_price
```

```python
import sort_dataframeby_monthorweek as sd         # for sorting monthwise

def sort_month(df, column):
    return sd.Sort_Dataframeby_Month(df, column)


sorted_total_hotels_price = sort_month(total_hotels_price, 'month')
sorted_total_hotels_price
```

```python
plt.figure(figsize = (10, 8))

plt.plot(sorted_total_hotels_price['month'], sorted_total_hotels_price['resort_hotel_price'], label='resort_hotel')
plt.plot(sorted_total_hotels_price['month'], sorted_total_hotels_price['city_hotel_price'], label='city_hotel')

plt.xlabel("Months")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.title("Avgerage room price per night over year")
plt.legend()
plt.show()
```

#### 2. Which are the most busy months?

```python
resort_hotel_guests = df_resort['arrival_date_month'].value_counts().reset_index()
resort_hotel_guests.columns=['month','No.of guests']
resort_hotel_guests
```

```python
city_hotel_guests = df_city['arrival_date_month'].value_counts().reset_index()
city_hotel_guests.columns=['month','No.of guests']
city_hotel_guests
```

```python
total_guests = resort_hotel_guests.merge(city_hotel_guests,on='month')
total_guests.columns=['month','No.of guests in resort hotel','No.of guests in city hotel']
total_guests
```

```python
sorted_total_guests = sort_month(total_guests,'month')
sorted_total_guests
```

```python
plt.figure(figsize = (10, 8))

plt.plot(sorted_total_guests['month'], sorted_total_guests['No.of guests in resort hotel'], label='No.of guests in resort hotel')
plt.plot(sorted_total_guests['month'], sorted_total_guests['No.of guests in city hotel'], label='No.of guests in city hotel')

plt.xlabel("Months")
plt.ylabel("No.of guests")
plt.xticks(rotation=45)
plt.title("Total guests per month")
plt.legend()
plt.show()
```

### Model Building

```python
y = df['is_canceled']
X = df.drop('is_canceled',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, random_state = 21, test_size = 0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
LR = LogisticRegression()

steps = [('standard_scaler', scaler), ('logistic_regression', LR)]
pipeline = Pipeline(steps)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
pipeline.score(X_test, y_test)
```

```python
final_data = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
final_data
```
